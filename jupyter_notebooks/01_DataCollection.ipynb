## Load Dataset 1 — Digital Marketing Campaign

```python
df1 = pd.read_csv('inputs/datasets/raw/digital_marketing_campaign_dataset.csv')
print(f"Shape: {df1.shape}")
df1.head()
```

```python
df1.info()
```

```python
missing = df1.isnull().sum()
print("Missing values per column:")
print(missing[missing > 0] if missing.sum() > 0 else "No missing values found.")
```

```python
print("Target variable — Conversion:")
print(df1['Conversion'].value_counts())
print(f"\nConversion rate: {df1['Conversion'].mean():.2%}")
```

```python
for col in ['AdvertisingPlatform', 'AdvertisingTool']:
    print(f"\n{col} unique values: {df1[col].unique()}")
```

## Conclusions

**Dataset 1 — Digital Marketing Campaign:**
- 8,000 records, 20 columns, no missing values
- Target `Conversion` is imbalanced: 87.65% positive class — SMOTE will be needed
- `AdvertisingPlatform` and `AdvertisingTool` contain only confidential placeholders — will be dropped
- `ConversionRate` is a derived metric that risks data leakage — will be excluded from ML features

**Dataset 2 — Marketing KPIs:**
- 308 records, 11 columns, no missing values
- 4 campaign categories: social (168), search (56), influencer (56), media (28)
- `campaign_name` has inconsistent capitalisation (e.g. facebOOK_tier2) — will be standardised in cleaning
- `c_date` is stored as object — will be parsed as datetime in cleaning

Both datasets are confirmed present and valid. Ready for Data Cleaning notebook.
